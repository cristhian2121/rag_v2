```python
from langchain_core.chat_history import BaseChatMessageHistory, InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
```

```python
### Load model
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# Load environment variables from .env file
load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
if api_key:
    print("✓ OPENAI_API_KEY loaded from .env")
else:
    print("⚠ Warning: OPENAI_API_KEY not found in environment")

model = init_chat_model(model="gpt-4o-mini", temperature=0.1)
```

```python
store = {}

def get_store(key: str) -> BaseChatMessageHistory:
    if key not in store:
        store[key] = InMemoryChatMessageHistory()
    return store[key]
```

```python
with_message_history = RunnableWithMessageHistory(model, get_store)
```

```python
config = {"configurable": {"session_id": "user1"}}
```

```python
response = with_message_history.invoke([
    ("user", "Cual e smi nombre")
], config)
```
