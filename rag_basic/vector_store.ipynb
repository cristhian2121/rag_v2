```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document


embedding_function = OpenAIEmbeddings()
```

```python

vector_store = Chroma(
    collection_name="test",
    embedding_function=embedding_function,
    persist_directory="./chroma_db",
)

documents = [
    Document(page_content="Hello, world!", metadata={"source": "example.txt"}),
    Document(page_content="This is a test!", metadata={"source": "example.txt"}),
    Document(page_content="This is another test!", metadata={"source": "example.txt"}),
    Document(page_content="This is yet another test!", metadata={"source": "example.txt"}),
]

vector_store.add_documents(documents)

```

```python
vector_store.similarity_search("Hello, world!", 2)
```

```python
vector_store.similarity_search_with_score("Hello, world!", 2)
```
